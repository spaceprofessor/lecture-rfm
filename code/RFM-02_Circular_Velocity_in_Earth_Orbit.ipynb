```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Earth:
    mu = 398600.4  # Gravitational parameter in [km^3/s^2]
    R = 6378.14  # Radius in [km]
```

## What is the velocity in a circular Earth orbit for a given altitude?

Relation between orbit altitude $h$ and orbit radius $r$:
\begin{equation*}
r = R + h
\end{equation*}

Velocity $V$ in a circular orbit with radius $r$:
\begin{equation*}
V = \sqrt\frac{\mu}{r} = \sqrt\frac{\mu}{R+h}
\end{equation*}

\begin{align*}
\mu &: \text{Earth's gravitational parameter}\\
R &: \text{Earth's radius}
\end{align*}

```python
h = 0  # Altitude in [km]
```

```python
e = Earth()

r = e.R + h
V = np.sqrt(e.mu/r)

print("The velocity in a circular Earth orbit with an altitude of", h, "km is", np.around(V, decimals=4), "km/s.")
```

## How does the velocity in a circular Earth orbit depend on altitude?

```python
plt.rcParams["figure.figsize"] = (12,8)
```

```python
h = np.arange(0, 1200)
r = e.R + h
V = np.sqrt(e.mu/r)
plt.plot(h, V)
plt.grid()
plt.xlabel("Altitude [km]")
plt.ylabel("Velocity [km/s]")
plt.show()
```

Note that circular Earth orbits at very low altitudes are not possible due to atmospheric drag.

## How does the orbital period of a circular Earth orbit depend on radius?

Orbital period $P$ of a circular orbit with radius $r$:
\begin{equation*}
P = 2\pi \sqrt\frac{r^3}{\mu}
\end{equation*}

```python
r = np.arange(0, 45000)
P = 2 * np.pi * np.sqrt(r**3/e.mu) / 3600
plt.plot(r, P)
plt.grid()
plt.xlabel("Orbit Radius [km]")
plt.ylabel("Orbit Period [h]")
plt.show()
```

## ... or on altitude?

Orbital period $P$ of a circular orbit with altitude $h$:
\begin{equation*}
P = 2\pi \sqrt\frac{(R+h)^3}{\mu}
\end{equation*}

```python
h = np.arange(0, 1200)
r = e.R + h
P = 2 * np.pi * np.sqrt(r**3/e.mu) / 60
plt.plot(h, P)
plt.grid()
plt.xlabel("Altitude [km]")
plt.ylabel("Orbit Period [min]")
plt.show()
```
